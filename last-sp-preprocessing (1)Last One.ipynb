```python
import csv

def replace_words(input_file, output_file, replacements):
    with open(input_file, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        rows = list(reader)
    
    # Replace words in the first row
    for i in range(len(rows[0])):
        for word, replacement in replacements.items():
            rows[0][i] = rows[0][i].replace(word, replacement)
    
    # Replace words in the second column
    for row in rows[1:]:
        for word, replacement in replacements.items():
            row[1] = row[1].replace(word, replacement)
    
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(rows)

# Define the input and output filenames
input_filename = '/kaggle/input/new-data-labels-last/New data labels.csv'
output_filename = 'AfterReplaceNewData.csv'

# Define the replacements as a dictionary
replacements = {
    'anger': 'غضب',
    'sadness': 'حزن',
    'Neutral': 'محايد',
    'fear': 'خوف',
    'happiness': 'سعادة',
    'disgust': 'اشمئزاز',
    'sarcasm': 'سخرية',
    'suprise': 'مفاجأة',
    'text': 'نص',
    'label': 'تصنيف',
    # Add more word replacements as needed
}

# Call the function to replace words
replace_words(input_filename, output_filename, replacements)

print("Replacement completed. Please check the output file:", output_filename)
```

## Removing diacritics

```python
import csv
import re

def get_text_with_arabic_diacritics(file_path):
    texts_with_diacritics = []
    with open(file_path, 'r', encoding='utf-8') as csv_file:
        csv_reader = csv.reader(csv_file)
        # Define a regex pattern to match Arabic diacritics
        arabic_diacritics_pattern = re.compile("[\u064B-\u065F\u0670\u0615-\u061A]")
        for row in csv_reader:
            for field in row:
                # Search for Arabic diacritics in each field
                match = arabic_diacritics_pattern.search(field)
                if match:
                    texts_with_diacritics.append(field)
    return texts_with_diacritics

# Example usage:
file_path = "/kaggle/working/AfterReplaceNewData.csv"
texts_with_diacritics = get_text_with_arabic_diacritics(file_path)
if texts_with_diacritics:
    print("Texts containing Arabic diacritics:")
    for text in texts_with_diacritics:
        print(text)
else:
    print("No text containing Arabic diacritics found in the CSV file.")
```

```python
import csv
import re

def remove_arabic_diacritics_from_csv(input_file_path, output_file_path):
    arabic_diacritics_pattern = re.compile("[\u064B-\u065F\u0670\u0615-\u061A]")
    
    with open(input_file_path, 'r', encoding='utf-8') as csv_input_file:
        csv_reader = csv.reader(csv_input_file)
        with open(output_file_path, 'w', encoding='utf-8', newline='') as csv_output_file:
            csv_writer = csv.writer(csv_output_file)
            for row in csv_reader:
                original_row = row.copy()  # Make a copy of the original row
                for i, field in enumerate(row):
                    # Remove Arabic diacritics from each field
                    row[i] = arabic_diacritics_pattern.sub('', field)
                csv_writer.writerow(row)

# Example usage:
input_file_path = "/kaggle/working/AfterReplaceNewData.csv"
output_file_path = "Removing_diacritics.csv"
remove_arabic_diacritics_from_csv(input_file_path, output_file_path)
print("Arabic diacritics removed from the text column in the CSV file.")
```

```python
import csv
import re

def print_original_and_modified_sentences(input_file_path, modified_file_path):
    arabic_diacritics_pattern = re.compile("[\u064B-\u065F\u0670\u0615-\u061A]")
    
    with open(input_file_path, 'r', encoding='utf-8') as input_csv_file, \
         open(modified_file_path, 'r', encoding='utf-8') as modified_csv_file:
        input_csv_reader = csv.reader(input_csv_file)
        modified_csv_reader = csv.reader(modified_csv_file)
        
        for original_row, modified_row in zip(input_csv_reader, modified_csv_reader):
            original_sentences = []
            modified_sentences = []
            
            for original_field, modified_field in zip(original_row, modified_row):
                # Check if the field in the original row contains Arabic diacritics
                if arabic_diacritics_pattern.search(original_field):
                    original_sentences.append(original_field)
                    modified_sentences.append(modified_field)
            
            if original_sentences:
                print("Original Sentences:", original_sentences)
                print("Modified Sentences:", modified_sentences)
                print()

# Example usage:
input_file_path = "/kaggle/working/AfterReplaceNewData.csv"
modified_file_path = "/kaggle/working/Removing_diacritics.csv"
print_original_and_modified_sentences(input_file_path, modified_file_path)
```

# Removing The English Caracters

```python
import re

def remove_english_characters(text):
    english_pattern = re.compile(r'[\dA-Za-z…]')
    cleaned_text = re.sub(english_pattern, '', text)
    return cleaned_text

# Specify the path to the input and output .txt files
input_file = "Removing_diacritics.csv"
output_file = "output_removeEnglishCar.csv"

# Read the Arabic text from the input file
with open(input_file, 'r', encoding='utf-8') as file:
    arabic_text = file.read()

# Remove English characters from the Arabic text
cleaned_arabic_text = remove_english_characters(arabic_text)

# Write the cleaned Arabic text to the output file
with open(output_file, 'w', encoding='utf-8') as file:
    file.write(cleaned_arabic_text)

print("English characters removed from the Arabic text and saved to the output file.")
```

# Removing Punctuations

```python
import re
import pandas as pd

def remove_english_characters(text):
    english_pattern = re.compile(r'[#@&%$!\(\)\{\}[\]:;.,،<>?؟`´''""\"''"""\[\]=+/\\|`''°~^*_-]')
    cleaned_text = re.sub(english_pattern, ' ', str(text))
    return cleaned_text

# Specify the path to the input and output CSV files
input_file = "output_removeEnglishCar.csv"
output_file = "output_remove_P_text.csv"

# Read the Arabic text from the input CSV file
df = pd.read_csv(input_file)

# Convert the 'نص' column to string if necessary
if df['نص'].dtype != object:
    df['نص'] = df['نص'].astype(str)

# Remove English characters from the Arabic text
df['نص'] = df['نص'].apply(remove_english_characters)

# Write the cleaned Arabic text to the output CSV file
#output_df = pd.DataFrame({'نص': cleaned_arabic_text})
df.to_csv(output_file, index=False)

print("English characters removed from the Arabic text and saved to the output CSV file.")
```

```python
import pandas as pd

# Load the emoji translations from the CSV file with proper encoding
emoji_data = pd.read_csv('//kaggle/input/tee000/tee.csv', encoding='utf-8')

# Create a dictionary from emoji to words
emoji_dict = dict(zip(emoji_data['emoji'], emoji_data['Meaning']))

def translate_emojis(text):
    translated_text = ""
    for char in text:
        if char in emoji_dict:
            translated_text += emoji_dict[char] + " "
        else:
            translated_text += char
    return translated_text

# Read input text from a file
with open('output_remove_P_text.csv', 'r', encoding='utf-8') as file:
    input_text = file.read()

# Translate emojis in the input text
translated_text = translate_emojis(input_text)

# Write translated text to an output file
with open('/kaggle/working/output_Remove_Emojis.csv', 'w', encoding='utf-8') as output_file:
    output_file.write(translated_text)

print("Emoji translation completed. Translated text saved to 'output_Remove_Emojis.txt' file.")
```

```python
import csv

def replace_words(input_file, output_file, replacements):
    with open(input_file, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        rows = list(reader)
        
    # Replace words in the first row
    for i in range(len(rows[0])):
        for word, replacement in replacements.items():
            rows[0][i] = rows[0][i].replace(word, replacement)
    
    # Replace words in the second column
    for row in rows[1:]:
        for word, replacement in replacements.items():
            row[1] = row[1].replace(word, replacement)
    
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(rows)

# Define the input and output filenames
input_filename = '/kaggle/working/output_Remove_ST.csv'
output_filename = '/kaggle/working/Finaloutput.csv'

# Define the replacements as a dictionary
replacements = {
    'غضب': 'anger',
    'حزن': 'sadness',
    'محايد': 'Neutral',
    'خوف': 'fear',
    'سعادة': 'happiness',
    'اشمئزاز': 'disgust',
    'سخرية': 'sarcasm',
    'مفاجأة': 'suprise',
    'نص': 'text',
    'تصنيف': 'label',
    # Add more word replacements as needed
}

# Call the function to replace words
replace_words(input_filename, output_filename, replacements)

print("Replacement completed. Please check the output file:", output_filename)
```

```python
import pandas as pd

def csv_to_excel(input_file, output_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(input_file)
    
    # Convert DataFrame to Excel file
    df.to_excel(output_file, index=False)

# Define the input and output filenames
input_filename = '/kaggle/working/Finaloutput.csv'
output_filename = 'output.xlsx'

# Call the function to convert CSV to Excel
csv_to_excel(input_filename, output_filename)

print("Conversion completed. Please check the output file:", output_filename)
```
